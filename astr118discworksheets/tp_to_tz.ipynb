```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

df = pd.read_csv("test.pt", sep="\s+", skiprows=1, header=None)
pressure = np.flip(df[1]) * 1e6
temperature = np.flip(df[2])

def pressureplot(x):
    plt.plot(x, pressure)
    plt.yscale('log')
    plt.gca().invert_yaxis()
    plt.ylabel(r"Pressure (erg/cm${}^3$)")
```

```python
pressureplot(temperature)
plt.xlabel("Temperature (K)")
```
